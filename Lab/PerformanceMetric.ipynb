BALANCED DATASET - Dataset is balanced → accuracy is reliable

Precision and recall support the accuracy value

Model performance is reasonably consistent#

```python
import tensorflow as tf

# Balanced dataset
y_true = tf.constant([1, 0, 1, 0, 1, 0, 1, 0], dtype=tf.float32)
y_pred = tf.constant([1, 0, 1, 1, 0, 0, 1, 0], dtype=tf.float32)

# Metrics
accuracy = tf.keras.metrics.BinaryAccuracy()
precision = tf.keras.metrics.Precision()
recall = tf.keras.metrics.Recall()

# Update metrics
accuracy.update_state(y_true, y_pred)
precision.update_state(y_true, y_pred)
recall.update_state(y_true, y_pred)

# Print results
print("Balanced Dataset Results")
print("Accuracy :", accuracy.result().numpy())
print("Precision:", precision.result().numpy())
print("Recall   :", recall.result().numpy())
```

IMBALANCED DATASET -

95% → class 0

5% → class 1

Model predicts all zeros

```python
# Imbalanced dataset
y_true = tf.constant([0]*95 + [1]*5, dtype=tf.float32)

# Bad model: predicts only majority class
y_pred = tf.constant([0]*100, dtype=tf.float32)

# Metrics
accuracy = tf.keras.metrics.BinaryAccuracy()
precision = tf.keras.metrics.Precision()
recall = tf.keras.metrics.Recall()

# Update metrics
accuracy.update_state(y_true, y_pred)
precision.update_state(y_true, y_pred)
recall.update_state(y_true, y_pred)

# Print results
print("\nImbalanced Dataset Results")
print("Accuracy :", accuracy.result().numpy())
print("Precision:", precision.result().numpy())
print("Recall   :", recall.result().numpy())
```
