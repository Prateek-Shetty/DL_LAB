```python
import matplotlib.pyplot as plt
import numpy as np

# Function and its derivative
# In CNN, x is one weight in filter, x**2 is loss function, 2x is gradient

def f(x):
    return x**2

def grad_f(x):
    return 2*x

# Gradient Descent settings
x = 8.0                 # starting point (far from minimum)
learning_rate = 0.1
iterations = 25

x_history = []

for i in range(iterations):
    x_history.append(x)
    gradient = grad_f(x)
    x = x - learning_rate * gradient
    print(f"Step {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")

# Plot
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label="f(x) = x²")
plt.scatter(x_history, f(np.array(x_history)), color='red')
plt.title("Gradient Descent: Rolling Down the Curve")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
